```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Se desea conocer como segun la temperatura cual es la ganancia de los helados
df = pd.read_csv('IceCreamSales-temperatures.csv')
# Muestra el DataFrame
df.head()
```

```python
plt.plot(df["Ice Cream Profits"],df["Temperature"],"b.")
plt.ylabel("Ice Cream Profits $USD")
plt.xlabel("Temperature F°")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Constr Modelo y ajuste funcion hipotesis
lin_reg = LinearRegression()
#Variable independiente (lo que me dice lo que quiero saber) variable dependiente (lo que quiero saber).
lin_reg.fit(df['Temperature'].values.reshape(-1,1),df['Ice Cream Profits'].values)
#lin_reg.fit(df['Ice Cream Profits'].values.reshape(-1,1),df['Temperature'].values)
```

```python
#theta 0
lin_reg.intercept_
```

```python
#theta 1
lin_reg.coef_
```

```python
#predic para el valor min y max del conjunto de datos
#X_min_max = np.array([[df["Ice Cream Profits"].min()],[df["Ice Cream Profits"].max()]])
X_min_max = np.array([[df["Temperature"].min()],[df["Temperature"].max()]])
y_train_pred = lin_reg.predict(X_min_max)
```

```python
plt.plot(X_min_max,y_train_pred,"g-")
#plt.plot(df["Ice Cream Profits"],df["Temperature"],"b.")
plt.plot(df["Temperature"],df["Ice Cream Profits"],"b.")
plt.ylabel("Ice Cream Profits $USD")
plt.xlabel("Temperature F°")
plt.show()
```

```python
x_new = np.array([[70]])
val = lin_reg.predict(x_new)
print(int(val[0]))
```

```python
plt.plot(X_min_max,y_train_pred,"g-")
#plt.plot(df["Ice Cream Profits"],df["Temperature"],"b.")
plt.plot(df["Temperature"],df["Ice Cream Profits"],"b.")
plt.plot(x_new,val,"rx")
plt.ylabel("Ice Cream Profits $USD")
plt.xlabel("Temperature F°")
plt.show()
```
