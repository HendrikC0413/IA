### Lectura de los datos.
Primero que nada vamos a importar los datos en un pandas, el cual en este caso voy a utilizar un data set de personas con el peso  y la altura.

```python
# Importamos las librerías necesarias para el análisis y leemos el documento csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('Clustering_gmm (2).csv')
data.head()
```

### Primera visualización de los datos.
Para continuar veremos como se comportan estas datos en una primera instancia.

```python
plt.scatter(data['Weight'], data['Height'])

plt.title('Gráfico de dispersión')
plt.xlabel('PESO')
plt.ylabel('ALTURA')

plt.show()
```

Aunque tiene forma lineal, se pueden observar 4 grupos claros que pueden deberse a que la persona tiene sobrepeso, sin embargo hay cuatro grupos claves.
### Normalizacion
Ahora aunque aparentemente no tiene outlayers aparentes vamos a aplicar la normalizacion por si exisitiese algun dato extraño en el data set. Y posteriormente observaremos un grafico para ver cambios con respecto al primer grafico de dispersion

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Datos_Normalizados = scaler.fit_transform(data)
plt.scatter(Datos_Normalizados[:, 0], Datos_Normalizados[:, 1])

plt.title('Gráfico de dispersión de datos normalizados')
plt.xlabel('PESO')
plt.ylabel('ALTURA')

plt.show()
```

De manera superficial no apreciamos muchos cambios mas que el cambio de los valores como tal
### Utilizacion de DBSCAN.
Ahora aplicaremos el dbscan donde debemos darle los parametros y luego relizar pedirle que realice la aplicacion y posteriormente lo visualizamos

```python
from sklearn.cluster import DBSCAN

# Ajustar los parámetros eps y min_samples
dbscan = DBSCAN(eps=0.5, min_samples=10)

# Aplica el algoritmo DBSCAN a los datos normalizados
clusters = dbscan.fit_predict(Datos_Normalizados)
```

```python
# Ahora, 'clusters' es un array que indica el cluster al que pertenece cada punto de tus datos
clusters
```

```python
plt.scatter(Datos_Normalizados[:, 0], Datos_Normalizados[:, 1], c=clusters)

plt.title('Gráfico de dispersión de clusters')
plt.xlabel('PESO')
plt.ylabel('ALTURA')

plt.show()
```

### Resultados primera instancia
Como se puede apreciar con los valores "Normales" que le puse a la esp (distancia entre los puntos) y la min_samples (numero minimo de muestras en la vecindad), Solo aparece un grupo destacado por amarillo y otro grupo de ruido.
